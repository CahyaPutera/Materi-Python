```python
import pandas as pd
import numpy as np
```

```python
# for standardize / normalisasi data
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import Binarizer
```

```python
# open data
df = pd.read_csv('melb_data.csv')
df.head()
```

```python
df_test = df[['Price', 'Distance', 'Car', 'Landsize', 'Rooms', 'Bedroom2', 'Bathroom']]
df_test.head()
```

### Standard Scaler

Digunakan pada data yang berdistribusi normal dan jika data cenderung tidak normal, Standard Scaler akan memaksa data berdistribusi normal.

```python
SScaler = StandardScaler()
Prize_scl = SScaler.fit_transform(df_test[['Price']])
df_test['Prize Std Scl'] = Prize_scl
```

```python
df_test.head()
```

```python
df_test['Price Std Scl 2'] = StandardScaler().fit_transform(df_test[['Price']])
```

```python
df_test.head()
```

### MinMax Scaler

Minmax Scaler digunakan pada kondisi feature yang tidak ada outliers dan dipakai sebagai cara pertama dalam proses Standardize feature. Bekerja dengan baik jika datanya tidak ada outlier. Karena MinMax Scaler sensitif dengan outlier. MinMax Scaler tidak mengintervensi data agar berdistribusi normal hanya merubah scale-nya saja.

```python
df_test['Price MinMax Sc'] = MinMaxScaler().fit_transform(df_test[['Price']])
```

```python
df_test.head()
```

### MaxAbs Scaler

MaxAbs dipakai ketika datanya tidak ada outlier

```python
df_test['Price MaxAbs Scl'] = MaxAbsScaler().fit_transform(df_test[['Price']])
```

```python
df_test.head()
```

### Robust Scaler

Robust Scaler di pakai jika feature memiliki data *Outlier*, karena scaler ini *tidak mendasarkan rumusnya dengan mean atau standar deviasi*. Scaler ini dasar teorinya menggunakan *IQR* sehingga lebih tahan (robust) dengan adanya outlier.

```python
df_test['Price Robust Scl'] = RobustScaler().fit_transform(df_test[['Price']])
```

```python
df_test.head()
```

### Normalizer

Sebenarnya untuk row

```python
df_test['Price Norm'] = Normalizer().fit_transform(df_test[['Price']])
```
