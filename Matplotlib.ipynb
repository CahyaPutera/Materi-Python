```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.figure(figsize = (12,5)) #size kanvasnya (panjang, lebar)

plt.plot(x,y,'r') #array, series, numerical di posisi x, dan y ,'r' = red

plt.xlabel('This is x', fontsize =10) #memberi nama X & ukuran font
plt.xticks([1,2,3,4,5], ['Bart', 'Lisa', 'Jack', 'Andy', 'As']) #memberi value X dan nama

plt.ylabel('This is y', fontsize =10) #memberi nama Y & ukuran font
plt.yticks([1,2,3,4,5]) #memberi value Y

plt.title('This is Title') #memberi title

plt.show() #memunculkan plot
```

```python
plt.plot(x, y,'r--') #membuat plot default
plt.plot(y, x,'g*')
```

```python
plt.figure(figsize = (12,8))
#subplot 
plt.subplot(2,2,1) #(row,col,urutan) (urutan ke 1)
plt.plot(x,y,'r--')
plt.title('Ini subplot 1')
plt.axis('off') #menghilangkan kanvas

plt.subplot(2,2,2) #(urutan ke 2)
plt.plot(x,y,'g*')
plt.title('Ini subplot 2')

plt.subplot(2,2,3) #(urutan ke 3)
plt.plot(x,y)
```

```python
fig = plt.figure(facecolor = 'yellow') #set warna kanvas

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #(posisi x, posisi y, panjang, lebar) dalam persen 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('Ini x1')
axes1.set_ylabel('Ini y1')
axes1.set_title('Ini title 1')

axes2.plot(x,y,'r')
axes2.set_xlabel('Ini x2')
axes2.set_ylabel('Ini y2')
axes2.set_title('Ini title 2')
axes2.set_axis_off()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (8,8), facecolor = 'yellow') #nrows = jumlah row, ncols = jumlah col, figsize = size kanvas
# for ax in axes:
#     ax.plot(x,y,'b')
#     ax.set_xlabel('Ini x')
#     ax.set_ylabel('Ini y')
#     ax.set_title('Ini title')
# axes

axes[0][0].plot(x,y,'b', label = 'Ini 1') #[row1][row2]
axes[0][0].plot(x,y, '--r', label = 'Ini 2')
axes[0][0].legend() #membuat legend
plt.tight_layout
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (8,8), facecolor = 'yellow') #nrows = jumlah row, ncols = jumlah col, figsize = size kanvas
for axe in axes:
    for ax in axe:
        ax.plot(x,y,'b')
        ax.set_xlabel('Ini x')
        ax.set_ylabel('Ini y')
        ax.set_title('Ini title')
 #looping   
```

```python
from yahoo_historical import Fetcher
```

```python
dfINDF = Fetcher('INDF.JK', [2011,1,1]).get_historical()
dfINDF
```

```python
dfINDF.head()
```

```python
dfINDF.isna().any()
```

```python
dfINDF.info()
```

```python
dfINDF = dfINDF.fillna(method = 'ffill')
```

```python
dfINDF.info()
```

```python
dfINDF['Date'] = pd.to_datetime(dfINDF['Date'] )
```

```python
dfINDF.info()
```

```python
(dfINDF['Date'][0] - dfINDF['Date'][2]).days
```

```python
dfINDF.info()
```

```python
dfINDF.set_index('Date', inplace = True)
```

```python
dfINDF
```

```python
plt.figure(figsize = (8,8))
plt.plot(dfINDF['Close'])
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 1, figsize = (25,30))
axes[0].plot(dfINDF['Open'], color = 'r')
axes[0].set_title('INDF Open', fontsize = 20)
axes[0].set_xlabel('Year', fontsize = 15)
axes[0].set_ylabel('Rupiah', fontsize = 15)
axes[1].plot(dfINDF['High'], color = 'g')
axes[1].set_title('INDF High', fontsize = 20)
axes[1].set_xlabel('Year', fontsize = 15)
axes[1].set_ylabel('Rupiah', fontsize = 15)
axes[2].plot(dfINDF['Volume'], color = 'b')
axes[2].set_title('INDF Volume', fontsize = 20)
axes[2].set_xlabel('Year', fontsize = 15)
axes[2].set_ylabel('Rupiah', fontsize = 15)
axes[3].plot(dfINDF['Adj Close'], color = 'y')
axes[3].set_xlabel('Year', fontsize = 15)
axes[3].set_ylabel('Rupiah', fontsize = 15)
axes[3].set_title('Adj Close', fontsize = 20)
plt.tight_layout
```
