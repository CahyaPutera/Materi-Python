```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.corr() #korelasi pearson (dlm bentuk matrix)
```

### Heatmap

```python
sns.heatmap(tips.corr(), annot = True)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
pvflight = flights.pivot_table(values = 'passengers', index = 'month', columns = 'year')
```

```python
pvflight
```

```python
plt.figure(figsize = (8,8))
sns.heatmap(pvflight) #heatmap harus dlm bentuk matrix
```

### Clustermap

```python
sns.clustermap(pvflight) #clusterin data algomerative (menggunakan tree based)
```

```python
sns.clustermap(pvflight, cmap = 'magma', linecolor = 'white', linewidths = 1)
```

### Grid

```python
tips
```

```python
sns.PairGrid(tips) #plot kosong
```

```python
g = sns.PairGrid(tips)
g.map(plt.scatter) #mapping apa yg dimasukkan ke plot kosong
```

```python
sns.PairGrid(tips).map(plt.scatter)
```

```python
g = sns.PairGrid(tips)
g.map_diag(sns.distplot, hist = False) #diagonal 
g.map_upper(sns.kdeplot) #atas
g.map_lower(plt.scatter) #bawah
```

### Facet Grid

```python
g = sns.FacetGrid(tips, col = 'time', row = 'smoker')
```

```python
g = sns.FacetGrid(tips, col = 'time', row = 'smoker', hue = 'sex') #membagi kategorical
g.map(plt.scatter, 'tip', 'total_bill').add_legend() #(plot(x,y))
```

### Joint Grid

```python
g = sns.JointGrid(x = 'total_bill', y = 'tip', data = tips)
g.plot(plt.scatter, sns.distplot) #satu axis 2 plot yg berbeda
```
